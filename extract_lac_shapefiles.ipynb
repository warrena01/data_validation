```python
import geopandas as gpd
```

```python
shapefile = gpd.read_file('my_path')
```

```python
shapefile
```

```python
# list of countries in LAC
my_list = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela', 'Mexico', 'Cuba', 'Dominican Republic', 'Haiti', 'Jamaica', 'Puerto Rico', 'Trinidad and Tobago', 'Bahamas', 'Barbados', 'Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 'Nicaragua', 'Panama', 'Antigua and Barbuda', 'Dominica', 'Grenada', 'Saint Kitts and Nevis', 'Saint Lucia', 'Saint Vincent and the Grenadines']
# countries in the gdf in LAC
my_countries = shapefile.loc[shapefile['COUNTRY'].isin(my_list)]
# update the my_list variable for countries in LAC & available in the gdf
my_list = my_countries['COUNTRY'].unique().tolist()
```

```python
import tkinter 
from tkinter import filedialog
dir_path = None
def get_file_directory():
    global dir_path
    root = tkinter.Tk()
    dir_path = filedialog.askdirectory()
    root.withdraw()       
get_file_directory()
print(dir_path)
```

```python
import os
```

```python
for each_country in my_list:
    country_gdf = shapefile.loc[shapefile['COUNTRY'] == each_country]
    destination_folder = dir_path + rf'/{each_country}'
    destination_file = destination_folder + rf'/{each_country}.shp'
    os.makedirs(destination_folder)
    country_gdf.to_file(destination_file)
```
